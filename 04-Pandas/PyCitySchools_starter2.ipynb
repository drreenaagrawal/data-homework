```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate the total number of schools: first find the number of unique schools, then find how many
unique_schools = school_data_complete["school_name"].unique()
total_schools = len(unique_schools)
total_schools
```

```python
#calculate total number of students by counting student ids
total_students = school_data_complete["Student ID"].count()
total_students
```

```python
#calculate total budget across all schools by summing the individual budgets of schools
total_budget = school_data["budget"].sum()
total_budget
```

```python
#average math score
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
#average reading score
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
#average score
average_score = (average_math_score + average_reading_score)/2
average_score
```

```python
#Calculate the percentage of students with a passing math score (70 or greater)
math_pass = school_data_complete.loc[school_data_complete["math_score"] >= 70]
math_pass_count = len(math_pass)
math_pass_percent = (math_pass_count*100)/total_students
math_pass_percent
```

```python
#Calculate the percentage of students with a passing reading score (70 or greater)
reading_pass = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
reading_pass_count = len(reading_pass)
reading_pass_percent = (reading_pass_count*100)/total_students
reading_pass_percent
```

```python
#summary table with the expected results
summary_table = pd.DataFrame(
{
   "Total Schools": [total_schools],
    "Total Students": [total_students],
    "Total Budget": [total_budget],
    "Average Math Score": [average_math_score],
    "Average Reading Score": [average_reading_score],
    "% Passing Math": [math_pass_percent],
    "% Passing Reading": [reading_pass_percent],
    "% Overall Passing Rate": [average_score],
}
)
summary_table
```

## Top Performing Schools (By Passing Rate)

* Sort and display the top five schools in overall passing rate

```python
#group data by school name
grouped_school_df = school_data_complete.groupby(["school_name"])
grouped_school_df.count().head()
```

```python
#total number of students in each school
total_students = grouped_school_df["Student ID"].count()
df_total_students = pd.DataFrame(total_students)
df_total_students.head()
```

```python
#the total budget for each school
total_school_budget = grouped_school_df["budget"].mean()
df_total_school_budget = pd.DataFrame(total_school_budget)
df_total_school_budget.head()
```

```python
#calculate per student budget
per_student_budget = df_total_school_budget["budget"]/df_total_students["Student ID"]
df_per_student_budget = pd.DataFrame(per_student_budget)
df_per_student_budget.head()
```

```python
#calculate average math score for each school
average_math_score_school = round(grouped_school_df["math_score"].mean(),2)
df_average_math_score_school = pd.DataFrame(average_math_score_school)
df_average_math_score_school.head()
```

```python
#calculate average reading score for each school
average_reading_score_school = round(grouped_school_df["reading_score"].mean(),2)
df_average_reading_score_school = pd.DataFrame(average_reading_score_school)
df_average_reading_score_school.head()
```

```python
#merging total students, total school budget, per student budget, avergae math score and averahe reading score
merge1 = pd.concat([df_total_students, df_total_school_budget, df_per_student_budget, df_average_math_score_school, df_average_reading_score_school ], axis=1)
merge1.head()
```

```python
#counting number of students per school who passed in math
math_pass_school = school_data_complete[school_data_complete["math_score"] >= 70]
math_pass_grouped = math_pass_school.groupby(["school_name"])
math_pass_grouped_count = math_pass_grouped["math_score"].count()
math_pass_grouped_count.head()
```

```python
#percent of students in each school that passed math
math_passpercent_by_school = math_pass_grouped_count * 100 / total_students
df_math_passpercent_by_school = pd.DataFrame(math_passpercent_by_school)
df_math_passpercent_by_school.head()
```

```python
#merging all the above statistics (merge 1 and % passing math)
merge2 = pd.merge(merge1, df_math_passpercent_by_school, on=["school_name"])
merge2.head()
```

```python
#renaming columns of the second merge to change the"0s"
merge2_renamed = merge2.rename(columns={
    "Student ID": "Total Students",
    "budget": "Total School Budget",
    "0_x": "Per Student Budget",
    "math_score": "Average Math Score",
    "reading_score": "Average Reading Score",
    "0_y": "% Passing Math"
})
```

```python
#counting number of students per school who passed in reading
reading_pass_school = school_data_complete[school_data_complete["reading_score"] >= 70]
reading_pass_grouped = reading_pass_school.groupby(["school_name"])
reading_pass_grouped_count = reading_pass_grouped["reading_score"].count()
reading_pass_grouped_count.head()
```

```python
#percent of students in each school that passed reading
reading_passpercent_by_school = reading_pass_grouped_count * 100 / total_students
df_reading_passpercent_by_school = pd.DataFrame(reading_passpercent_by_school)
df_reading_passpercent_by_school.head()
```

```python
#merge6 = pd.merge(merge5r, df_reading_passpercent_by_school, on=["school_name"])
```

```python
overall_passing_rate = (reading_passpercent_by_school + math_passpercent_by_school)/2
df_overall_passing_rate = pd.DataFrame(overall_passing_rate)
df_overall_passing_rate.head()
```

```python
merge3 = pd.merge(df_reading_passpercent_by_school, df_overall_passing_rate, on=["school_name"])
merge3_renamed = merge3.rename(columns={
    "0_x": "% Passing English",
    "0_y": "% Overall Passing Rate"
})
merge3_renamed.head()
```

```python
school_type = grouped_school_df["type"].first()
df_school_type = pd.DataFrame(school_type)
df_school_type_renamed = df_school_type.rename(columns={
    "type": "School Type"
})
df_school_type_renamed.head()
```

```python
#merging all results
final = pd.concat([df_school_type_renamed, merge2_renamed, merge3_renamed], axis=1)
final_df = pd.DataFrame(final)
final_df
```

```python
#top 5 best performing schools
best_performers_df = final_df.sort_values("% Overall Passing Rate", ascending = False)
best_performers_df.head(5)
```
